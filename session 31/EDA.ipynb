```python
numerical_columns = train_data.select_dtypes(["float64", "int64"]).columns
```

```python
fig = plt.figure(figsize=(22,20))

fig.add_axes([0,0,1,1])
ax = fig.get_axes()[0]

corr_mat = train_data[numerical_columns].corr()


sns.heatmap(corr_mat, ax=ax, vmin=-1, vmax=1, annot=True)
plt.savefig("./heatmap.jpg", bbox_inches="tight")
plt.show()
```
